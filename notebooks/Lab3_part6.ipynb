```r
dat <- read.csv(file="June 10-July 12, 2015 - Gaming, Jobs and Broadband - CSV.csv", header=TRUE)
```

```r
myvars <- c("age", "sex", "intfreq", "game4","emplnw")
```

```r
newdf <- df[myvars]
```

```r
newdf <- na.omit(newdf)
```

```r
ndf <- newdf[ which(newdf$sex < 3 & newdf$intfreq < 6 & newdf$game4 < 3 & newdf$emplnw < 5), ]
```
